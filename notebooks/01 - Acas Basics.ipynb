# Test Performance of direct vs indirect attribute getting and setting

```python
from array import array

class Test1:
    def __init__(self, data):
        self.a, self.b, self.c, self.d, self.e = data
    def get_data(self):
        return self.a, self.b, self.c, self.d, self.e

class Test2:
    attributes = ("a","b","c","d","e")
    def __init__(self, data):
        for k, v in zip(self.attributes, data):
            setattr(self, k, v)
    def get_data(self):
        return (getattr(self, k) for k in self.attributes)

class Test3:
    attributes = ("a","b","c","d","e")
    def __init__(self, data):
        for k, v in zip(self.attributes, data):
            self.__dict__[k]=v
    def get_data(self):
        return (self.__dict__[k] for k in self.attributes)

class Test4(array):
    def __new__(cls, data):
        return array.__new__(cls, 'f', (data[:5]))
    def get_data(self):
        return self
    @property
    def a(self): return self[0]
    @property
    def b(self): return self[1]
    @property
    def c(self): return self[2]
    @property
    def d(self): return self[3]
    @property
    def e(self): return self[4]

```

```python
data = (1,2,3,4,5)
```

```python
%%timeit
o1 = Test1(data)
```

```python
%%timeit
o2 = Test2(data)
```

```python
%%timeit
o3 = Test3(data)
```

```python
%%timeit
o4 = Test4(data)
```

```python
o1 = Test1(data)
o2 = Test2(data)
o3 = Test3(data)
o4 = Test4(data)
```

```python
%%timeit
a,b,c,d,e = o1.get_data()
```

```python
%%timeit
a,b,c,d,e = o2.get_data()
```

```python
%%timeit
a,b,c,d,e = o3.get_data()
```

```python
%%timeit
a,b,c,d,e = o4.get_data()
```

```python
%%timeit
a,b,c,d,e = o1.a,o1.b,o1.c,o1.d,o1.e
```

```python
%%timeit
a,b,c,d,e = o4.a,o4.b,o4.c,o4.d,o4.e
```
